```python
# check for outliers int the population data

col_names = [
             'total_inactive_population', 'total_unemployed_population',
             'total_employed_population']

fig, ax = plt.subplots(len(col_names), figsize=(6, 6))  # Adjusted figsize
for i, col_val in enumerate(col_names):
    sns.boxplot(x=kenya_pop_df[col_val], ax=ax[i])
    ax[i].set_title('Boxplot - {}'.format(col_val), fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=8)

plt.tight_layout()
plt.show()
```

```python

# Column names for comparison
columns_to_compare = ['total_inactive_population', 'ILO_unemployed_rate',
                      'total_unemployed_population', 'total_employed_population']

# Plotting
plt.figure(figsize=(10, 6))

# Plot 'Information and communication' against each column
for col in columns_to_compare:
    plt.plot(kenya_pop_df['year'], kenya_pop_df['Information and communication'], label='Information and communication')
    plt.plot(kenya_pop_df['year'], kenya_pop_df[col], label=col)
    plt.title(f'Comparison of Information and communication with {col}')
    plt.xlabel('Year')
    plt.ylabel('Values from ' + col)  # Update y-axis label to include the column name
    plt.legend()
    plt.grid(True)
    plt.show()

```
